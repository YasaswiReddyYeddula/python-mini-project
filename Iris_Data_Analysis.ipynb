Data Analysis on Iris Dataset

*  Performing basic mathematical operation on Dataset



The **mean()** operation is used to find the average values

```python
#Find average sepal length.
Avg_sepal_length = df["sepal_length"].mean()
print("Average of sepal length :", Avg_sepal_length)
```

```python
#Find average petal length.
Avg_petal_length = df["petal_length"].mean()
print("Average of petal length :",Avg_petal_length)
```

The **max()** operation is used to find the maximum value

```python
#Find maximum sepal length.
max_sepal_length = df["sepal_length"].max()
print("Maximum sepal length :", max_sepal_length)
```

The **min()** operation is used to find the minimum value

```python
#Find minimum petal width.
min_petal_width = df["petal_width"].min()
print("Minimum petal length :",min_petal_width)
```

**nunique()** is used to find the unique values in a dataset

**len()** used to find the length

```python
#Find total number of flower species.
total_flower_species = df["species"].nunique()
print("Types of flower species:", total_flower_species)

flower_species = len(df["species"])
print("Total no.of flower species :", flower_species)
```

**value_counts()** returns the count of each unique value in a column.

```python
#Find count of each species.
species_count = df["species"].value_counts()
print(species_count)
```

**Visualization using Matplotlib**

**What is Matplotlib -**
   Matplotlib is a Python library used to create charts and graphs for data visualization.

Types of charts and graphs in matplotlib :
   

*   **Line Chart –** Shows trends over time or ordered data.

*   **Bar Chart –** Compares values across different categories.

*   **Pie Chart –** Shows the proportion of each category in a whole.

*   **Histogram –** Displays the distribution (frequency) of numerical data.

*   **Scatter Plot –** Shows the relationship between two numerical variables.

*   **Box Plot –** Displays the spread and outliers in the data.


```python
#CHART 1: LINE CHART
print("Line chart :")

plt.figure(figsize = (10,8))
plt.plot(df.index, df["sepal_length"], marker='o', color="Green")
plt.title("Sepal Length by index")
plt.xlabel("Index")
plt.ylabel("Sepal_length")
plt.grid(True)
Line_chart_Path = "Sepal_Lengt_Line_Chart.png"
plt.savefig(Line_chart_Path, bbox_inches = 'tight')
plt.show()
plt.close()

print("Line chart is saved as Sepal_Lengt_Line_Chart.png")
```

```python
#CHART 2: BAR CHART
species_count = df["species"].value_counts()

print("Bar chart :")

plt.figure(figsize = (10,8))
plt.bar(species_count.index, species_count.values, color="Skyblue")
plt.title("Count of Species")
plt.xlabel("Species")
plt.ylabel("Species Count")
Bar_chart_Path = "Species_count_Bar_Chart.png"
plt.savefig(Bar_chart_Path, bbox_inches = 'tight')
plt.show()
plt.close()

print("Bar chart is saved as Species_count_Bar_Chart.png")
```

```python
#CHART 3: SCATTER PLOT
print("Scatter plot :")

plt.figure(figsize = (10,8))
plt.scatter(df["sepal_length"], df["petal_length"], marker='o', color="Red")
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.grid(True)
Scatter_chart_Path = "SP_Length_scatter_Chart.png"
plt.savefig(Scatter_chart_Path, bbox_inches = 'tight')
plt.show()
plt.close()

print("Scatter plot is saved as SP_Length_scatter_Chart.png")
```

```python
#CHART 4: HISTOGRAM
print("Histogram :")

plt.figure(figsize = (10,8))
plt.hist(df["sepal_width"], bins = 20, color = "SkyBlue" )
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
Histogram_chart_Path = "Sepal_width_Histogram.png"
plt.savefig(Histogram_chart_Path, bbox_inches = 'tight')
plt.show()
plt.close()

print("Histogram is saved at:", Histogram_chart_Path)
```

**Visualization using Plotly**

**What is plotly -**  Plotly is a Python library used to create interactive and visually appealing charts and graphs for data visualization.

| **Matplotlib**                                   | **Plotly**                                          |
| ------------------------------------------------ | --------------------------------------------------- |
| Creates **static** charts.                       | Creates **interactive** charts.                     |
| Best for basic plots and reports.                | Best for dashboards and web applications.           |
| Less visually appealing by default.              | More modern and attractive by default.              |
| Requires more code for customization.            | Easier to create interactive charts with less code. |
| Charts cannot be zoomed or hovered (by default). | Supports zooming, panning, and hover tooltips.      |


```python
#Chart 1 : Interactive Scatter Plot
print("Scatter Plot : ")

Fig = px.scatter(
    df,
    x = "sepal_length",
    y = "petal_length",
    color = "species",
    title = "Sepal Length vs Petal Length_Scatter Plot"
)
Fig.show()
```

```python
#CHART 2 : Box Plot
print("Box Plot :")

bx = px.box(
    df,
    x = "species",
    y = "petal_length",
    color = "species",
    title = "Species vs Petal Length Box Plot"
)
bx.show()
```

```python
#CHART 3 : HISTOGRAM
print("Histogram :")

hist = px.histogram(
    df,
    x = "sepal_length",
    nbins = 20,
    title = "Distribution of sepal length"
)
hist.show()
```

```python
#CHART 4: PIE CHART
print("Pie Chart :")

species_count = df["species"].value_counts().reset_index()
species_count.columns = ["species", "count"]

pie_chart = px.pie(
    species_count,
    names = "species",
    values = "count",
    title = "Total Distribution of Species"
)
pie_chart.show()
```
